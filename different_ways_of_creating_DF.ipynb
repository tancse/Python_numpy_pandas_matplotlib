DIFFERENT WAYS OF CREATING DATAFRAME

1)USING CSV
2)USING EXCEL
3)FROM PYTHON DICTIONARY
4)FROM LIST OF TUPLES
5)FROM LIST OF DICTIONARIES

```python
import pandas as pd
```

# READ FROM CSV

```python
df = pd.read_csv("C:\\python_numpy_pandas_matplotlib\\Datasets\\nyc_weather.csv")
df
```

# CREATING THROUGH EXCEL

```python
df = pd.read_excel("C:\\python_numpy_pandas_matplotlib\\Datasets\\weather_data.xlsx", "Sheet1")
df
```

# CREATING USING DICTIONARY

```python
weather_data = {
    'date' : ['1/1/2017', '1/2/2017', '1/3/2017', '1/4/2017', '1/5/2017', '1/6/2017'],
    'temperature' : [32,35,28,24,32,31],
    'windspeed' : [6,7,2,7,4,2],
    'event' : ['Rain', 'Sunny', 'Snow', 'Snow', 'Rain', 'Sunny']
}
df = pd.DataFrame(weather_data)
print(df)
```

# CREATING USING TUPLES LIST

```python
weather_data =[
    ('1/1/2022',32,6,'rain'),
    ('1/2/2022',35,7,'sunny'),
    ('1/3/2022',28,2,'snow')
]
df = pd.DataFrame(weather_data, columns=["date","temp","windspeed","event"])
df
```

# CREATING DF USING LIST OF DICTIONARY

```python
weather_data = [
    {'date':'1/1/2017', 'temperature': 32, 'windspeed': 6, 'event': 'rain'},
    {'date':'1/2/2017', 'temperature': 35, 'windspeed': 4, 'event': 'rain'},
    {'date':'1/3/2017', 'temperature': 28, 'windspeed': 2, 'event': 'rain'}
]

df = pd.DataFrame(weather_data)
df
```
