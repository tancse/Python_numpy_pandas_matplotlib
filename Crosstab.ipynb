```python
import pandas as pd

df = pd.read_excel('survey.xls')
df
```

```python
#crosstabs are used to create contiguous table
df1 = pd.crosstab(df.Nationality,df.Handedness)
df1
```

```python
df1 = pd.crosstab(df.Sex,df.Handedness)
df1
```

```python
df1 = pd.crosstab(df.Sex,df.Handedness,margins=True)
df1
```

```python
df1 = pd.crosstab(df.Sex,[df.Handedness,df.Nationality],margins=True)
df1
```

```python
df1 = pd.crosstab([df.Sex,df.Nationality],df.Handedness,margins=True)
df1
```

```python
df1 = pd.crosstab(df.Sex,df.Handedness, margins=True)
df1
```

```python
df1 = pd.crosstab(df.Sex,df.Handedness, normalize='index')
df1
```

```python
import numpy as np
df1 = pd.crosstab(df.Sex,df.Handedness, values=df.Age, aggfunc=np.average)
df1
```
