NOW DOING READ AND WRITE FUNCTION USING EXCEL FILE

```python
df = pd.read_excel("Datasets/stock_data.xlsx","Sheet1")
df
```

```python
#to convert the NA to some value in excel
def convert_people_cell(cell):
    if cell=='n.a.':
        return 'sam walton'
    return cell

df = pd.read_excel("Datasets/stock_data.xlsx",
                   "Sheet1",
                   converters={
                       'people':convert_people_cell
                    }
                   )
df
```

WRITING INTO EXCEL FILE

```python
df.to_excel("new.xlsx",sheet_name="stocks")
# if we don't want index , give index = False
# to write the excel starting from column 2 and row 2
#give parameter startcol=2,startrow=1
```
