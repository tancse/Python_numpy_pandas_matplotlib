# PIVOT OR PIVOT_TABLE
# It converts the given table into some form where 
 # some of the columns are chnaged into the index and column.

```python
df.pivot(index="date",columns="city")
```

```python
df.pivot(index="date",columns="city",values='humidity')
```

```python
df.pivot(index="humidity",columns="city")
```

# PIVOT TABLE

Allows us to summarize and aggregate data inside dataframe


```python
df1 = pd.read_csv('Datasets/weather2.csv')
df1
```

```python
df1.pivot_table(index="city",columns="date")
# so the repeation of some dataset with different value in one single column returns average of the value
# that's what happend in the humidity and temperature table.
# which is known as aggregate function or aggfunc
```

```python
df1.pivot_table(index="city",columns="date",aggfunc="sum")
#aggfunc default value is average, but we can do sum also
```

```python
df1.pivot_table(index="city",columns="date",aggfunc="count")
```

```python
df1.pivot_table(index="city",columns="date",aggfunc="diff")
```

```python
df1.pivot_table(index="city",columns="date",aggfunc="max")
```

```python
df1.pivot_table(index="city",columns="date",margins=True)
```
