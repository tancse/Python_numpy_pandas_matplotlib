# Problem statement
    -mpg dataset
    -predict mpg for a given car sample

```python
mpg = sns.load_dataset('mpg')
```

```python
mpg = mpg.dropna()
```

```python
y = mpg.pop('mpg')
y
```

```python
X = mpg.drop(['origin','name'], axis=1)  #strings cannot be handled by machine learning throws error as 
                                         # str should be converted to float
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.25)
```

```python
len(Xtrain),len(Xtest)
```

```python
len(ytrain),len(ytest)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(Xtrain, ytrain)
```

# to handle missing values

```python
y_pred = lr.predict(Xtest)
y_pred
```

```python
ytest
```

```python
(y_pred == ytest).sum()
```
