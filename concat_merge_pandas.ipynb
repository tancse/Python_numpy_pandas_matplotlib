# MERGE ----> little AUTOMATION of CONCAT
# THE INDEX WHAT WE DID  IN THE ABOVE CONCAT IS AUTOMATED IN MERGER


```python
import pandas as pd
df1 = pd.DataFrame({
    'city':['mumbai','delhi','chennai','bangalore'],
    'temperature':[37,41,45,21]
})
df1
```

```python
df2 =pd.DataFrame({
    'city':['delhi','mumbai','chennai','hyderbad'],
    'humidity':[7,12,9,21]
})
df2
```

```python
# to join these df we can use merge (doing inner join - default)
df = pd.merge(df1,df2,on="city")
df
```

```python
# to do outer join
df = pd.merge(df1,df2,on="city",how="outer")
df
```

```python
# to do left join
df = pd.merge(df1,df2,on="city",how="left")
df
```

```python
# to do right join
df = pd.merge(df1,df2,on="city",how="right")
df
```

```python
# to do outer join - and get to know from which df it came from
df = pd.merge(df1,df2,on="city",how="outer",indicator=True)
df
```
