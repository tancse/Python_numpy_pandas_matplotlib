```python
import pandas as pd

weather_data = {
    'date' : ['1/1/2017', '1/4/2017', '1/5/2017', '1/6/2017','1/7/2017','1/8/2017','1/9/2017','1/10/2017','1/11/2017'],
    'temperature' : [32,' ' ,28, ' ',32, ' ',' ' ,34,40],
    'windspeed' : [6,9, ' ',7, ' ',' ' ,' ' ,8,12],
    'event' : ['Rain', 'Sunny', 'Snow',' ' , 'Rain', 'Sunny', ' ', 'Cloudy','Sunny']
}
df = pd.DataFrame(weather_data)
df.to_csv('weather_data.csv',index=False)
```

#STARTING TO FILL MISSING VALUES

fillna - to fill missing values using different ways
interpolate to make a guess on missing values using interpolation
dropna to drop rows with missing values

```python
df = pd.read_csv('weather_data.csv',
                 na_values=['not avaialbe', ' '])
df.to_csv('weather_data.csv',index=False)
df
```

```python
df = pd.read_csv('weather_data.csv', parse_dates=["date"])
type(df.date[0])
df
```

```python
df.set_index('date',inplace=True)
df
```

# FILLNA

```python
new_df=df.fillna(0) #we have to give the value we want
new_df
```

```python
#filna for different col with different values
new_df = df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'no event'
})
new_df
```

```python
#to fill different and senesible data into the file
new_df = df.fillna(method="ffill")
#ffill standards for forward fill which copies previous value to the new ccell
#so previous day wethear will be copied to the next day also.
new_df
```

```python
#also we can use bfill - backward fill reverse of ffill
new_df = df.fillna(method="bfill")
new_df
```

```python
new_df = df.fillna(method="bfill", axis="rows")
new_df
# AXIS has two values 'rows and columns' whihc fill the na in vertical or horizontal manner
```

```python
#LIMIT
new_df = df.fillna(method="ffill",limit=1)
#only copies to one cell not the next consequent cells
new_df
```

#still not happy with the fillna because inconsistency between data
# INTERPOLATE

```python
new_df=df.interpolate() #gives a reasonable data to fill
new_df
```

```python
new_df = df.interpolate(method="time")#now it is realizing the date near 4 is 5 and fillin wiht the near value to 5
new_df
```

# DROP

```python
new_df=df.dropna() #drops the missing values row.
new_df
```

```python
new_df = df.dropna(how="all") #drps only the row with all values are na.
new_df
```

```python
# how to drop if atleat one none na value keep it and drop others
new_df = df.dropna(thresh=2)
new_df
```
